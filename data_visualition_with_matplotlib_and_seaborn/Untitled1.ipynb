```python
# rengini şeklini degiştirme 
dizi_1 = np.random.randint(1000,5000,12)
dizi_2 = np.random.randint(1000,5000,12)

uzunluk = np.arange(dizi_1.shape[0])
plt.plot(dizi_1, "bo-",label="Dizi 1")
plt.plot(dizi_2,"r--",label="Dizi 2")

plt.xlabel("Aylar")
plt.ylabel("Fiyat/TL")

plt.title("2024 kripto Para Fiyatları")
plt.legend()

plt.xticks(uzunluk,["Ocak","Subat","Mart","Nisan","Mayıs","Haziran","Temmuz","Agustos","Eylül","Ekim","Kasım","Aralık"],rotation=45)
plt.show()
```

```python
# linestyle ise nasıl yazmamızı istersek yazar 
# "bo- " ve "r--" ise renginii gösterir 
dizi_1 = np.random.randint(1000,5000,12)
dizi_2 = np.random.randint(1000,5000,12)

uzunluk = np.arange(dizi_1.shape[0])
plt.plot(dizi_1,linestyle= ":",label="Dizi 1")
plt.plot(dizi_2, linestyle= "--",label="Dizi 2")

plt.xlabel("Aylar")
plt.ylabel("Fiyat/TL")

plt.title("2024 kripto Para Fiyatları")
plt.legend()

plt.xticks(uzunluk,["Ocak","Subat","Mart","Nisan","Mayıs","Haziran","Temmuz","Agustos","Eylül","Ekim","Kasım","Aralık"],rotation=45)
plt.show()
```

```python
fiyat_btc = np.random.randint(100,250,5)
fiyat_eth = np.random.randint(100,250,5)

Dizi_isim = ["Bitcoin", "Ethereum"]

ort_btc= np.mean(fiyat_btc)
ort_eth = np.mean(fiyat_eth)

new_dizi = [ort_btc, ort_eth]

plt.bar(Dizi_isim,new_dizi,color=["yellow","pink"],label=["BTC","ETH"])

plt.legend() 
for i in range(len(Dizi_isim)):
    plt.text(i,new_dizi[i],f"{new_dizi[i]:.2f}",ha="center",va="bottom")
plt.show()
```
