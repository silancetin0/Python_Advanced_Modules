```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
# DataFrame  oluşturma(kodlar)

stn_boy = np.random.randint(150,190,(10))
stn_kilo = np.random.randint(50,100,(10))
stn_no = np.random.randint(2100,2200,(10))

# pandas kütüphanesi, dictionary veri yapısı üzerine kurulur.
# pandas kütüphanesi ile excel ve sql'daki her şeyi yapabilir.

my_dict = {"ogr_no":stn_no,"ogr_boy":stn_boy,"ogr_kilo":stn_kilo}
my_df = pd.DataFrame.from_dict(my_dict)
my_df
```

```python
my_df.index
```

```python
my_df.index = ["ogrenci_1","ogrenci_2","ogrenci_3","ogrenci_4","ogrenci_5","ogrenci_6","ogrenci_7","ogrenci_8","ogrenci_9","ogrenci_10"]
my_df
```

```python
# DataFrame'ye yeni sütün ve değeler ekleme
stn_vize = np.random.randint(20,100,(10))
stn_final = np.random.randint(20,100,(10))

my_df["ogr_vize"] = stn_vize
my_df["ogr_final"] = stn_final
my_df
```

```python
deger_vize = my_df["ogr_vize"].values
deger_final = my_df["ogr_final"].values
```

```python
# ogr_gecti isimli bir key ekleyin
#bu key'in değerleri vize notlarının yüzde kırkı ile 
# final notlarının yüzde altısınının toplamı olacak 
```

```python
deger_vize = my_df["ogr_vize"].values*0.40 
deger_final= my_df["ogr_final"].values*0.60 
deger_gecti = deger_vize+ deger_final
my_df["ogr_gecti"]=deger_gecti
my_df
```

```python
my_df["durum"] = my_df["ogr_gecti"].values>50
my_df
```

```python
my_df["cinsiyet"] = "erkek"
my_df
```

```python
#istenilen özelliklere göre deger getirme
#tek sorguya göre 
my_df[my_df["ogr_vize"]<50]
```

```python
#birden fazla sorguya göre deger çagırma
my_df[(my_df["ogr_vize"]<50) & (my_df["durum"]== True) & (my_df["ogr_boy"]>170)]
```
